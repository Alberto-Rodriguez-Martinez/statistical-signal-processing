## 1. Modelling the World: From Problem to Solution

## 2. The Inference Cycle: Reality, Measurement, Processing, Result

We break down the process of gaining knowledge from the world into four fundamental stages:

    <img src="../images/inference_cycle.png" alt="The inference cycle" width="600" style="border: 2px solid #4CAF50;">
</div>

### 2.1. Reality (Nature)

This is where the phenomenon we wish to study resides. It is the underlying "truth."

*   **Hypothesis ($H_i$):** A supposition about reality (e.g., is there a signal or not?).
*   **Parameter ($\theta_i$):** Unknown numerical characteristics of reality (e.g., voltage, speed, amplitude, noise power).
  
### 2.2. What I Have (Measurement/Observation/Transmission)

We do not directly access reality. Instead, we measure or receive data that is a "dirty" and noisy version of reality. Our observation is affected by **uncertainty**.

*   **Data ($X$):** The measurements we obtain, often affected by noise or interference.
*   **Observation Model $f(x, \theta_i)$:** A mathematical description of how reality transforms into the observed data. **It is our *best guess* of how the phenomenon manifests in our measurements**, including the effect of noise. They are usually probability rules (probability density functions, pdf) that connect data (measurements) with the unknown parameters (estimates).

### 2.3. What I Do (Processing)

This is the stage where we apply algorithms and rules to the observed data to infer something about reality.

*   **Processing:** Processing the data (measurments and known parameters) to obtain useful information (filter to reduce noise, recuce complexity, correlation, etc.)
*   **Estimation Function:**  The algorithm that produce estimates of parameters ($\theta_i$).
*   **Decision Rule** The algorithms that transform the data ($X$) into decisions about hypotheses ($H_i$)

### 2.4. What I Get (Result)

The final product of our processing. It can be a discrete decision or a continuous value.

*   **Estimation/Estimate:** Assigning a continuous numerical value to an unknown parameter (e.g., an object's speed).
*   **Detection:** Deciding between a finite set of possibilities (e.g., "there is a signal" vs. "there is no signal").
*   **Classification:** Assigning an observation to one of several predefined categories (e.g., aircraft type, assign fingerprint to person).

And we would have to provide **ALWAYS!** information about the **performance** of the estimator and detector: 
* **Estimator:** Bias & variance.
* **Detector:** Probability of Detection, probability of False Alarm, probability of Error, ROC's, etc.

## 3. Goals and Methods of Statistical Signal Processing

### 3.1. Main Goal

**To infer useful information from messy or incomplete data.**

*   **Y:** Unknown data (what we want to know)
*   **X:** Known data (the observation)
*   **Z:** Unknown data (irrelevant, noise)

### 3.2. Strategy

Formulate a probabilistic model of the data ($x$), which depends on underlying parameter(s) ($\theta$).

*   **$\theta$:** Model parameters.
*   **$f(x|\theta)$:** Probability density function of the model.

### 3.3. Main Approaches

*   **Estimation Problems:** Assigning a continuous value to a parameter(s).
    *   Scalar: $\theta \in \mathbb{R}$
    *   Vector: $\theta \in \mathbb{R}^k$
*   **Detection Problems:** Deciding between a discrete set of options.
    *   Binary detection: $\theta \in \{0, 1\}$ (e.g., yes/no)
    *   Classification: $\theta \in \{1, ..., M\}$ (e.g., type A, B, C)

### 3.4. Key Tasks

1.  Create statistical models for measured data.
2.  Find fundamental limitations / Quality factors.
3.  Develop optimal (or suboptimal) estimators.
4.  Asymptotic analysis of estimator performance.
5.  Simulations and experiments.

### 3.5. Methods

*   Treat everything generally (unified mathematical representation).
*   Bias towards Gaussian noise and linear observation.
*   **Classical:** Hypotheses/parameters are fixed (deterministic), non-random.
*   **Bayesian:** Hypotheses/parameters are treated as random variables with assumed *a priori* distributions.

##### Example: 
I want to measure the level of a battery in an electronic device, so the user can change it before something wrong happens. The unkown parameter would be here the battery voltage V, and I will measure it using an alectronic circuit within the device. I would have to stablish a threshol, let say 20% of the nominal value of the battery (which is 1.5 V, threrefore the threshoold would be 0.3 V), so if the level of the batterry decreases below that value, a LED will turn red and remain like that until the user changes the battery. In this example, the parameter to be estimated is the battery voltage V, and we would have to make a detector to detect wether the level is below or above the stablished threshold.

<div style="text-align: center;">
    <img src="../images/battery_circuit_reduced.png" alt="Voltage meassurement example" width="200" style="border: 2px solid #4CAF50;">

```python
# Set constans
V_Nominal = 1.5 # Nomial value of the battery voltage
Threshold_per_cent = 20 # percent of the nominal value of the battery voltage
Threshold = V_Nominal * Threshold_per_cent / 100 # calculates the actual threshold to be used in the detector
V = 0.322 # real value of the voltage of the battery, deterministic, constant, not measurable...
```

We will have to define the model now. As I am an experienced engineer, I know that the measuring device will be affected by electronic noise, and that usually it can be modeled as an additive white gaussian noise, modelled as a Gaussian distribution $N(\mu, \sigma^2)$, with $\mu$ its mean, we will assume here it is zero, and mean power (variance) $\sigma^2$. we will deal later with these concepts, but for now just assume that this defines how the noise behaves.\
Assuming the above is true, our hipothesis would be that at any time, $X$, the measured signal (sample), will be the sum of the real value of the voltage plus some noise (uncertain/random value), that is, $X=V+n$, with $n$ the value of the noise, modeled as the wait gaussina random variable. It means that our measurement behaves as a Gaussian variable with mean $\mu_V=V+\mu$. Note that the voltage $V$ is deterministic, unknown, but deterministic.

So far, what have is:
* **Reality**:\
$V$, real voltage of the battery
* **Model**: Measurement, $X$, is affected by additive white gaussian noise\
$noise \sim N(\mu, \sigma^2)=\frac{1}{\sigma \sqrt{2\pi}} \, e^{-\frac{(x - \mu)^2}{2\sigma^2}}$\
$X=V+noise \sim N(\mu_V, \sigma^2)=\frac{1}{\sigma \sqrt{2\pi}} \, e^{-\frac{(x - \mu_V)^2}{2\sigma^2}}$\
$\mu_V=\mu+V$

* **Processing**: None so far...
* **Result**: Just $\widehat{V}=X$, the measured value...

```python
# Import libraries
import numpy as np

# We model the noise sample using python libraries that model random distributions
var = 0.01 # the variance, or mean power in case or white gaussian noise
std = np.sqrt(var) # note that the standard deviation can generally be calculated as the square root of the variance
mean = 0 # the mean
noise = np.random.normal(mean, std) # Gaussian with mean mu and standard deviation sd
X = V + noise # According to our model observed (measured) sample is the sum of the real value plus the noise

print(f"The real value of the voltage is {V:.4f}")
print(f"The measured voltage is {X:.4f}")
print(f"The noise is {noise:.4f}")
```

Now, we have to analyze if this is a good method to asses the battery voltage. If you remember anything from your classes of statistics 0.1, a gaussian distribution of mean $\mu_V$ and standard deviation $\sigma$, will produce values around the mean with certain probability according to the Gaussian distribution.

| Interval | Probabilit |
|----------|----------|
| $\mu_V+\sigma$ | $\approx 68.27\%$| 
| $\mu_V+2\sigma$ | $\approx 95.45\%$| 
| $\mu_V+3\sigma$ | $\approx 99.73\%$| 

This can be seen graphically if we plot the probability density function of the random variable.


```python
import matplotlib.pyplot as plt
from scipy.stats import norm

# define new mean. Note that the noise is additive gaussian and the unknown parameter V is deterministic
mean_V = V + mean
# range
x = np.linspace(mean_V - 4*std, mean_V + 4*std, 1000)

# create the probability density function (PDF)
pdf_X = norm.pdf(x, mean_V, std)

# plot
plt.plot(x, pdf_X, 'b', linewidth=1, label='Distribution of the measures - Gaussian PDF')
plt.axvline(mean_V, color='r', linestyle='--', linewidth=1, label=rf'$\mu_V$ = {mean_V}')

# Rellenar las áreas bajo la curva para ±1σ, ±2σ y ±3σ
for n, color, alpha in zip([1, 2, 3], ['green', 'orange', 'red'], [0.3, 0.2, 0.1]):
    x_fill = np.linspace(mean_V - n*std, mean_V + n*std, 500)
    y_fill = norm.pdf(x_fill, loc=mean_V, scale=std)
    plt.fill_between(x_fill, y_fill, color=color, alpha=alpha, label=rf'$\pm{n}\sigma$ region')

    # Flechas horizontales en cada σ
    y_sigma = norm.pdf(mean_V + n*std, loc=mean_V, scale=std)
    plt.annotate(
        '',
        xy=(mean_V - n*std, y_sigma),
        xytext=(mean_V + n*std, y_sigma),
        arrowprops=dict(arrowstyle='<->', lw=1.5, color=color)
    )
    # Texto a la derecha de la curva
    plt.text(mean_V + n*std + std * 0.2, y_sigma,
             rf'$\pm{n}\sigma=\pm{n*std:.3f}$', color=color,
             ha='left', va='center', fontsize=9)
    
# labels and format
plt.title('Gaussian Distribution')
plt.xlabel('Value')
plt.ylabel('Probability Density')
# Legend BELOW the figure
plt.legend(
    loc='upper center',
    bbox_to_anchor=(0.5, -0.15),
    fancybox=True,
    shadow=False,
    ncol=2
)
plt.grid(True, linestyle=':', linewidth=0.7)

plt.show()
```

If the noise power is for example $\sigma^2=0.01$, that means that the standard deviation is $\sigma=0.1$. Now, if the real voltage was $V=0.322$ volts, we will measure values within the following intervals woith the corresponding probabilities.

| Interval | Probability |
|----------|----------|
| $[0.272, 0.372]$ | $\approx 38.3\%$| 
| $[0.222, 0.422]$ | $\approx 68.27\%$| 
| $[0.122, 0.522]$ | $\approx 95.45\%$| 
| $[0.022, 0.622]$ | $\approx 99.73\%$| 

As you can see, this is not a very good estimator. We can calculate the probability of measuring values below the threshold and above the threshold for specific values of voltages, as shown in the next code cell. The figure shows the distribution, centered around the mean ($\mu_V$), of the possible observations $X$ for a given value of voltage $V$. The area to the left of the threshold line represents the probability that the measurement lays below the threshold, and to the right is the probability of measuring a value above the threshold. 

As our detector will activate the low battery LED once the level measured is below the threshold, and will remain there no matter what, we have a big problem, as even for low values of noise power the probability activating the LED when the real value is above the threshold is quiet high. Note that it will wrongly activate in the scenario of having a real voltage above the threshold and measuring a value below the threshold. There is another error scenario, in which the real value is below the threshold but the measure lays above the threshold. As you see there is a lot to consider when estimating and detecting random variables... 

In this curse we will learn how to analyse this sort of problems, how to deisgn estimators, detectors and clasifiers, and how to asses their quality, in terms of eficiency, reliabilyty, etc. 

You can play around with the value of the noise power and the real voltage to evaluate different escenarios.

```python
# Set constans
V_Nominal = 1.5 # Nomial value of the battery voltage
Threshold_per_cent = 20 # percent of the nominal value of the battery voltage
Threshold = V_Nominal * Threshold_per_cent / 100 # calculates the actual threshold to be used in the detector
V = 0.322 # real value of the voltage of the battery, deterministic, constant, not measurable...
var = 0.01 # the variance, or mean power in case or white gaussian noise
std = np.sqrt(var) # note that the standard deviation can generally be calculated as the square root of the variance
mean = 0 # the mean of the noise
mean_V = mean + V # the mean of the observations

# --- probability X ≤ Threshold ---
p_below = norm.cdf(Threshold, loc=mean_V, scale=std)
print(f"P(X ≤ {Threshold}) = {p_below:.4f}")

# --- range and PDF ---
x = np.linspace(mean_V - 4*std, mean_V + 4*std, 1000)
pdf = norm.pdf(x, loc=mean_V, scale=std)

# --- Plot PDF ---
plt.figure(figsize=(7, 4))
plt.plot(x, pdf, linewidth=2, color='navy', label='Gaussian PDF')

# --- threshold line ---
plt.axvline(Threshold, color='k', linestyle='--', linewidth=1.5, label=rf'Threshold $T={Threshold}$')

# --- fill areas ---
x_below = x[x <= Threshold]
plt.fill_between(x_below, norm.pdf(x_below, mean_V, std), color='green', alpha=0.35, label=r'$X ≤ Threshold$')

x_above = x[x >= Threshold]
plt.fill_between(x_above, norm.pdf(x_above, mean_V, std), color='red', alpha=0.20, label=r'$X > Threshold$')

# --- labels and format ---
#plt.title(rf'$P(X \\le Threshold) = {p_below:.3f}$  for  $X \sim \mathcal{{N}}(\mu={mean_V},\,\sigma={std})$')
plt.xlabel('Value')
plt.ylabel('Probability Density')
plt.legend(loc='upper center', bbox_to_anchor=(0.5, -0.18), ncol=3)
plt.grid(True, linestyle=':', linewidth=0.7)
plt.tight_layout()
plt.show()
```

Just to show some simple strategies we will analyse and design, in this example, we can just change the detector by measuring several samples of the voltage, lets say $N$ samples, and just making the average. This is the intuitive solution for such a simple problem, and we will demonstrate that this is an optimal solution, as it has the minimum poissible variance, which will result in $\sigma^2_V=\sigma^2 / N$.

If you see next figure, it seems very intuitive that the average will produce a better result than a simple sample. Next we plot the probability distributions of both measurement systems, and see how they behave around the threshold regardint the probability of making a wrong decission.

```python
# Set constans
V_Nominal = 1.5 # Nomial value of the battery voltage
Threshold_per_cent = 20 # percent of the nominal value of the battery voltage
Threshold = V_Nominal * Threshold_per_cent / 100 # calculates the actual threshold to be used in the detector
V = 0.322 # real value of the voltage of the battery, deterministic, constant, not measurable...
var = 0.01 # the variance, or mean power in case or white gaussian noise
std = np.sqrt(var) # note that the standard deviation can generally be calculated as the square root of the variance
mean = 0 # the mean of the noise
mean_V = mean + V # the mean of the observations

# take N samples
N = 10 # number of samples qe acquiere and average
x = V + np.random.normal(loc=mean, scale=std, size=10) # generate 10 samples of noise and ad the real voltage

# --- Plot ---
plt.figure(figsize=(7, 4))
plt.plot(x, 'o', color='navy', label='Measurements')
plt.plot(np.ones(np.size(x))*V, linewidth=1, color='navy', label='Real voltage')
plt.plot(np.ones(np.size(x))*np.mean(x), linewidth=1, color='coral', label='Average of the measurements')

plt.xlabel('samples')
plt.ylabel('Voltage')
plt.legend(loc='upper center', bbox_to_anchor=(0.5, -0.18), ncol=3)
plt.grid(True, linestyle=':', linewidth=0.7)
plt.tight_layout()
plt.show()
```

So now our scenario looks like this:

* **Reality**:\
$V$, real voltage of the battery
* **Model**: Measurement, $X$, is affected by additive white gaussian noise. 
$X=V+noise$\
$noise \sim N(\mu, \sigma^2)=\frac{1}{\sigma \sqrt{2\pi}} \, e^{-\frac{(x - \mu)^2}{2\sigma^2}}$\
$X\sim N(\mu_V, \sigma^2)=\frac{1}{\sigma \sqrt{2\pi}} \, e^{-\frac{(x - \mu_V)^2}{2\sigma^2}}$\
$\mu_V=\mu+V$

* **Processing**: We take $N$ measurements $\underline{x}=\left\{ x(1), x(2), ..., x(N) \right\}$ and average them to produce the estimate.\
$x(i)=V+noise(i)$\
$noise(i) \sim N(\mu, \sigma^2)=\frac{1}{\sigma \sqrt{2\pi}} \, e^{-\frac{(x - \mu)^2}{2\sigma^2}}$\
$\overline{X}=\frac{1}{N}\sum_{i=1}^{N}x(i)$\
$\overline{X}\sim N(\mu_V, \sigma^2_V)=\frac{1}{\sigma_V \sqrt{2\pi}} \, e^{-\frac{(x - \mu_V)^2}{2\sigma^2_V}}$\
$\mu_V=\mu+V, \sigma^2_V=\sigma^2/N$
* **Result**:
    * **Estimate:** $\widehat{V}=\overline{X}=\frac{1}{N}\sum_{i=0}^{N}x(i)$
    * **Detection:** if $\widehat{V} > Threshold \rightarrow Led On!$

There is a big improvement here. If you check the standar deviation (square root of the variance), it has decreased noy by a factor $\sqrt N$, which means that the interval confidence of the averaged measure is much better, therefore the measure will be more reliable (and accurate in mean).

Next figure shows how the probabilites of error change when we take the average of several samples. You can try changing the number of samples $N$ and see how it could affect the performance of the detector.

```python
# calculate the standar deviation of the average
# take N samples
N = 20 # number of samples qe acquiere and average
std_V = std/np.sqrt(N)


# --- probability X ≤ Threshold ---
p_below = norm.cdf(Threshold, loc=mean_V, scale=std)
p_below_av = norm.cdf(Threshold, loc=mean_V, scale=std_V)
print(f"For a single measurement P(X ≤ {Threshold}) = {p_below:.4f}")
print(f"For the average of {N} measurements P(X ≤ {Threshold}) = {p_below_av:.4f}")

# --- range and PDF ---
x = np.linspace(mean_V - 4*std, mean_V + 4*std, 1000)
pdf = norm.pdf(x, loc=mean_V, scale=std)
pdf_av = norm.pdf(x, loc=mean_V, scale=std_V)

# --- Plot PDF ---
plt.figure(figsize=(7, 4))
plt.plot(x, pdf, linewidth=2, color='navy', label='PDF one measurmeent')
plt.plot(x, pdf_av, linewidth=2, color='red', label=rf'PDF of the average of {N} measurements')

# --- threshold line ---
plt.axvline(Threshold, color='k', linestyle='--', linewidth=1.5, label=rf'Threshold $T={Threshold}$')

# --- labels and format ---
#plt.title(rf'$P(X \\le Threshold) = {p_below:.3f}$  for  $X \sim \mathcal{{N}}(\mu={mean_V},\,\sigma={std})$')
plt.xlabel('Value')
plt.ylabel('Probability Density')
plt.legend(loc='upper center', bbox_to_anchor=(0.5, -0.18), ncol=3)
plt.grid(True, linestyle=':', linewidth=0.7)
plt.tight_layout()
plt.show()
```
